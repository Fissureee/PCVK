```python
import numpy as np
import matplotlib.pyplot as plt

# Load the image (assuming balloon.jpg is in grayscale format)
filename = '/content/drive/MyDrive/PCVK_assets/balloon.jpg'
img = plt.imread(filename)

# If the image is in RGB, convert it to grayscale
if img.ndim == 3:
    img = np.dot(img[...,:3], [0.299, 0.587, 0.114]).astype(np.uint8)

# Step 1: Calculate histogram
histogram, bin_edges = np.histogram(img, bins=256, range=(0, 256))

# Step 2: Calculate probabilities for each intensity level
pixel_counts = img.size
probabilities = histogram / pixel_counts

# Step 3: Iterate over all possible thresholds to find the optimal one
max_variance = 0
optimal_threshold = 0

for threshold in range(256):
    # Background (0 to threshold)
    weight_bg = np.sum(probabilities[:threshold])
    if weight_bg == 0:
        continue

    mean_bg = np.sum([i * probabilities[i] for i in range(threshold)]) / weight_bg

    # Foreground (threshold to 255)
    weight_fg = np.sum(probabilities[threshold:])
    if weight_fg == 0:
        continue

    mean_fg = np.sum([i * probabilities[i] for i in range(threshold, 256)]) / weight_fg

    # Inter-class variance
    variance_between = weight_bg * weight_fg * (mean_bg - mean_fg) ** 2

    # Update the optimal threshold if the variance is the maximum found so far
    if variance_between > max_variance:
        max_variance = variance_between
        optimal_threshold = threshold

# Step 4: Apply the optimal threshold to get the binary image
binary_otsu = np.where(img > optimal_threshold, 255, 0)
# Global threshold for comparison (using a fixed threshold, e.g., 127)
global_threshold = 90
binary_global = np.where(img > global_threshold, 255, 0)

# Display the images and threshold values
titles = [f'Original Image', f'Global Threshold (T={global_threshold})', f"Otsu's Threshold (T={optimal_threshold})"]
images = [img, binary_global, binary_otsu]

plt.figure(figsize=(15, 5))
for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()

print(f"Otsu's optimal threshold value: {optimal_threshold}")
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the color image
filename = '/content/drive/MyDrive/PCVK_assets/tobacco.jpg'
img_color = cv.imread(filename)

# Convert to grayscale for thresholding
img_gray = cv.cvtColor(img_color, cv.COLOR_BGR2GRAY)

# Apply adaptive thresholding to create a mask for better segmentation
mask = cv.adaptiveThreshold(img_gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C,
                            cv.THRESH_BINARY_INV, 11, 2)

# Use morphological operations to clean up the mask
kernel = np.ones((5,5), np.uint8)
mask = cv.morphologyEx(mask, cv.MORPH_CLOSE, kernel)

# Apply mask to show only the foreground
foreground = cv.bitwise_and(img_color, img_color, mask=mask)
background = cv.bitwise_and(img_color, img_color, mask=cv.bitwise_not(mask))

# Calculate histograms for the full image (entire image)
hist_full = [cv.calcHist([img_color], [i], None, [256], [0, 256]) for i in range(3)]

# Calculate histograms only for the foreground using the mask
hist_foreground = [cv.calcHist([img_color], [i], mask, [256], [0, 256]) for i in range(3)]

# Plotting the results
plt.figure(figsize=(10, 10))

# Show the original color image
plt.subplot(2, 2, 1)
plt.imshow(cv.cvtColor(img_color, cv.COLOR_BGR2RGB))
plt.title("Citra Asli")
plt.axis('off')

# Show the segmentation mask
plt.subplot(2, 2, 2)
plt.imshow(mask, cmap='gray')
plt.title("Citra Segment")
plt.axis('off')

# Display histogram for the full image
plt.subplot(2, 2, 3)
colors = ('b', 'g', 'r')
for i, color in enumerate(colors):
    plt.plot(hist_full[i], color=color)
plt.title("Histogram dari keseluruhan citra")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

# Display histogram only for the foreground
plt.subplot(2, 2, 4)
for i, color in enumerate(colors):
    plt.plot(hist_foreground[i], color=color)
plt.title("Histogram hanya pada FG objek saja")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
